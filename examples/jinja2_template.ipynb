### with pandas DataFrame

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame(np.random.rand(5, 5))
df2 = pd.DataFrame(np.random.rand(5, 5))
df1_html = df1.to_html()
df2_html = df2.to_html()
```

```python
%%tsumiki -r

:HTML::
{{ df1_html }}

:HTML::
{{ df2_html }}
```

#### with matplotlib

```python
from io import StringIO
import matplotlib.pyplot as plt

svg_data = StringIO()

fig, ax = plt.subplots()
ax.plot(df1)
fig.savefig(svg_data, format="SVG")
plt.close(fig)
svg_data.seek(0)

plot_data = svg_data.read()
```

```python
%%tsumiki -r

:Markdown:
# DataFrame and Plot

:HTML::
{{ df1_html }}
:HTML::
{{ plot_data }}
```
